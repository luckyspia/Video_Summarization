```python
/key
    /features                 2D-array with shape (n_steps, feature-dimension)
    /gtscore                  1D-array with shape (n_steps), stores ground truth importance score (used for training, e.g. regression loss)
    /user_summary             2D-array with shape (num_users, n_frames), each row is a binary vector (used for test)
    /change_points            2D-array with shape (num_segments, 2), each row stores indices of a segment
    /n_frame_per_seg          1D-array with shape (num_segments), indicates number of frames in each segment
    /n_frames                 number of frames in original video
    /picks                    positions of subsampled frames in original video
    /n_steps                  number of subsampled frames
    /gtsummary                1D-array with shape (n_steps), ground truth summary provided by user (used for training, e.g. maximum likelihood)
    /video_name (optional)    original video name, only available for SumMe dataset
```

```python
# PySceneDetect
# 프레임 단위 전환점 자동 탐색
# !pip install scenedetect[opencv]
from scenedetect import VideoManager, SceneManager, StatsManager
from scenedetect.detectors import ContentDetector
from scenedetect.scene_manager import save_images, write_scene_list_html
```

```python
video_path = './data/AIhub/Training/SUMVID_MEDIUM_TRAIN_01/0be7a3bcb2cf2b779387d24aac721e9c83b62a993e851b51ba46fb38c5f16507-유튜브 영상물(5~20분) 장면별 중요도 태그-100177-20201217181222-001-001.mp4'
stats_path = 'result.csv' # 영상의 각 프레임마다 전환점에 대한 점수를 나타냄

video_manager = VideoManager([video_path])
stats_manager = StatsManager()
scene_manager = SceneManager(stats_manager)

scene_manager.add_detector(ContentDetector(threshold=30)) # 0~100 까지, 낮을수록 민감하게 변화에 반응
                           # 가까운 프레임 간의 색상과 밝기 등을 비교해서 전환점을 탐색

video_manager.set_downscale_factor() # 처리 속도 향상을 위해 이미지 크기를 자동으로 줄임

video_manager.start()
scene_manager.detect_scenes(frame_source=video_manager)

# result
with open(stats_path, 'w') as f:
    stats_manager.save_to_csv(f, video_manager.get_base_timecode())

scene_list = scene_manager.get_scene_list()
print(f'{len(scene_list)} scenes detected!')

# 전환점 시작 장면으로 썸네일 추출
save_images(
    scene_list,
    video_manager,
    num_images=1,
    image_name_template='$SCENE_NUMBER',
    output_dir = 'scenes'
)

# write_scene_list_html('result.html', scene_list)

for scene in scene_list:
    start, end = scene
    print(f'{start.get_seconds()}-{end.get_seconds()}') 
```
